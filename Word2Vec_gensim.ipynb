```python
from tensorflow import keras
import os
import gensim
```

```python
from gensim.models import KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin',binary=True)
```

```python
#check size of each word vector
model.vector_size
```

```python
model.most_similar('Delhi',topn=1)
```

```python
result = model.most_similar(positive=['man','queen'],negative=['king'],topn=3)  #topn=3
print(result)
```

```python
result = model.most_similar(positive=['Delhi','France'],negative=['India'],topn=1)
print(result)
```

```python
result = model.most_similar(positive=['He'],negative=['Queen'],topn=1)
print(result)
```

```python
print(model.similarity('Delhi','India'))
```
